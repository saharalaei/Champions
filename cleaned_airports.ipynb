```python
import pandas as pd
```

```python
airports = pd.read_csv("airports.csv")
airports_df = pd.DataFrame(airports)
airports_df
```

```python
airports_df = airports_df.rename(columns = {"id":"ID", "name":"Name","ident":"Code", "type":"Type", "latitude_deg":"Latitude", "longitude_deg":"Longitude", "continent":"Continent", "iso_country": "Country", "iso_region":"Region", "municipality":"City"})
airports_df = airports_df[["ID", "Name", "Code", "Type", "Latitude", "Longitude", "Continent", "Country", "Region", "City"]]
airports_df
```

```python
airports_df.to_csv("cleaned_airports.csv")
```
