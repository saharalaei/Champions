```python
import pandas as pd 
```

```python
covids = pd.read_csv('owid-covid-data (1).csv')
covids_df = pd.DataFrame(covids)
covids_df
```

```python
covids_df = covids_df.rename(columns ={"iso_code":"Code", "continent":"Continent","location":"Country",
                                        "date":"Date"})
covids_df = covids_df[["Code", "Continent","Country", "Date","total_cases","new_cases",
                       "total_deaths","new_deaths","total_cases_per_million","new_cases_per_million"]]
covids_df
```

```python
covids_df.to_csv("cleaned_covid.csv")
```
